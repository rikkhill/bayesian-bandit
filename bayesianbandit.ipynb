```python
import pymc3 as pm
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# beta/binomial model

n = 30
vals = np.random.binomial(n=n, p=0.00976, size=2000)
```

```python

with pm.Model() as model:
    a = pm.Uniform("alpha", 1, 100)
    b = pm.Uniform("beta", 1, 100)
    p = pm.Beta("p", alpha=a, beta=b)
    binom = pm.Binomial("binom", n=n, p=p, observed=vals)
    trace1 = pm.sample(5000, tune=2000)
```

```python
pm.traceplot(trace1, varnames=["p"])
```

```python
pm.plot_posterior(trace1)
```
